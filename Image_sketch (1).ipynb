#Import Required Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

#Importing Image

```python
#Original Image
df=cv2.imread("/content/sachin.jpeg")
cv2_imshow(df)
```

```python
#Grey Scale Image
grey_df=cv2.cvtColor(df,cv2.COLOR_BGR2GRAY) 
cv2_imshow(grey_df)
```

```python
#Inverting the grey scale image
inverted_df = cv2.bitwise_not(grey_df)  
cv2_imshow(inverted_df)
```

```python
#image smoothing
df_smoothing = cv2.GaussianBlur(inverted_df, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(df_smoothing)
```

#Pencil sketch

```python
def pencil_sketch(x, y):

    return cv2.divide(x, 255 - y, scale=256)
```

```python
#pencil sketch
pencil_sketch=pencil_sketch(grey_df,df_smoothing)
cv2_imshow(pencil_sketch)
```

#Showing Original Image and the Pencil Sketch

```python
cv2_imshow(df)
cv2_imshow(pencil_sketch)
```
